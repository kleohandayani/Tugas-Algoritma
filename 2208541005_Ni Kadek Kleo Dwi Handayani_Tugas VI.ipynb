# **5.21.4 Mengubah data pada Pandas**

```python
import pandas as pd
```

```python
data = pd.read_csv('bunga_iris.txt')
print(data)
```

```python
#perintah ini adalah pengubahan data dengan melakukan operasi fungsi
#data yang akan diubah diinput dalam variabel yang diinginkan
#pada perintah ini juga akan mencetak kolom baru sesuai dengan variabel string yang dibuat

data['tiga kali sepal length (cm)'] = 3*data['sepal length (cm)']
print(data)
```

```python
#perintah ini digunakan untuk menghapus kolom yang kita inginkan

data = data.drop(columns=['tiga kali sepal length (cm)'])
print(data)
```

```python
#cell ini membuat perintah pengubahan data dengan menjumlahkan variabel pada semua baris dan kolom 2-4
#disini juga akan dibuatkan kolom baru yang dimana merupakan hasil dari operasi sum

data['Jumlah'] = data.iloc [:,2:4].sum(axis=1)
print(data)
```

```python
#perintah ini digunakan untuk menghapus kolom yang kita inginkan

data = data.drop(columns = ['Jumlah'])
print(data)
```

# **5.21.7 Pengelompokan data menggunakan Pandas**

```python
import pandas as pd
```

```python
data = pd.read_csv('bunga_iris.txt')
print(data)
```

```python
#perintah ini dapat mengelompokkan data yang dimana menggunakan fungsi mean()
#fungsi mean() digunakan untuk menghitung rata-rata dari sebuah kolom

data.groupby(['sepal length (cm)']).mean()
```

```python
#'sort_values' digunakan untuk mengurutkan nilai terkecil ke terbesar
#'ascending=True' ini membantu 'sort_values' dalam mengurutkan nilai

data.groupby(['sepal length (cm)']).mean().sort_values('sepal width (cm)', ascending=True)
```

```python
#fungsi 'sum()' digunakan untuk menghitung jumlah dari sebuah kolom

data.groupby(['sepal length (cm)']).sum()
```

```python
#fungsi 'count()' digunakan untuk menghitung jumlah kolom dan baris

data.groupby(['sepal length (cm)']).count()
```

# **5.22.1 Grafik sederhana dengan Matplotlib**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('bunga_iris.txt')
print(data)
```

```python
x = data['sepal length (cm)']
y = data['petal length (cm)']
plt.plot(x,y)
plt.title('Grafik Sepal dan Petal')
plt.xlabel('X = panjang sepal')
plt.ylabel('Y = panjang petal')
```

# **5.22.2 Properti Pembuatan Grafik dengan Matplotlib**

```python
x = data['sepal length (cm)']
y = data['petal length (cm)']
x1 = data['sepal width (cm)']
y1 = data['petal width (cm)']
plt.figure(figsize=(10,10), dpi=400)
plt.plot(x,y,label = 'Grafik length',color='red',linewidth = 3,marker='.',markersize=15, 
         markeredgecolor='blue',linestyle='-')
plt.plot(x1,y1,label= 'Grafik width',color ='yellow',linewidth=3,marker='.',markersize=15, 
         markeredgecolor = 'green')
plt.title('Grafik Sepal dan Petal', fontdict={'fontname':'DejaVu Sans','fontsize':20})
plt.xlabel('X', fontdict={'fontname':'Arial','fontsize':6})
plt.ylabel('Y', fontdict={'fontname':'Arial','fontsize':6})
plt.xticks(['sepal length (cm)'])
plt.yticks(['petal length (cm)'])
plt.legend()
plt.savefig('grafik.png', dpi=400)
plt.show()
```

# **5.22.3 Histogram sederhana dengan Matplotlib**

```python
sepal = data['sepal length (cm)']
petal = data['petal length (cm)']
sepal1 = data['sepal width (cm)']
petal1 = data['petal width (cm)']
plt.xlabel('Sepal (cm)')
plt.ylabel('Petal (cm)')
print(plt.bar(sepal,sepal1))
print(plt.bar(petal,petal1))
```
